```python

hoten = input("Ten chu ho: ")
so_truoc = int(input("Chi so thang truoc: "))
so_sau = int(input("Chi so thang nay: "))
kwh = so_sau - so_truoc
tien_dien = 0
if kwh <= 50:
    tien_dien = kwh * 1984
elif kwh <= 100:
    tien_dien = 50 * 1984 + (kwh - 50) * 2050
elif kwh <= 200:
    tien_dien = 50 * 1984 + 50 * 2050 + (kwh - 100) * 2380
elif kwh <= 300:
    tien_dien = 50 * 1984 + 50 * 2050 + 100 * 2380 + (kwh - 200) * 2998
elif kwh <= 400:
    tien_dien = 50 * 1984 + 50 * 2050 + 100 * 2380 + 100 * 2998 + (kwh - 300) * 3350
else:
    tien_dien = 50 * 1984 + 50 * 2050 + 100 * 2380 + 100 * 2998 + 100 * 3350 + (kwh - 400) * 3460
tong_tien = tien_dien * 1.08
print("Ho va ten:", hoten)
print("Tien phai tra la:", round(tong_tien))
```
