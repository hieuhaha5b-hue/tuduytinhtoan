Bài 18:Xoay mảng sang phải x bước.
Cho danh sách a có độ dài n và số nguyên không âm x, hãy viết hàm
rotate_right(a, x) trả về một danh sách mới là kết quả của việc xoay a sang phải x vị trí.
Ví dụ, a = [5, 2, 1, 4, 3] và x = 2 → [4, 3, 5, 2, 1].
- Không thay đổi danh sách đầu vào.
- Không dùng tuple hoặc dictionary

```python
def rotate_right(a, x):
    n=len(a)
    x=x%n
    if x==0:
        return a
    else:
        return a[-x:]+a[:-x]
a=list(map(int,input().split()))
x=int(input())
print(rotate_right(a, x))
```

Bài 27:Viết hàm is_perfect(n) trả về True nếu n là số hoàn hảo và False nếu không phải. Một số nguyên dương được gọi là hoàn hảo nếu nó bằng
tổng các ước dương thực sự của nó (không tính chính nó). Theo định nghĩa, n phải dương; với giá trị không dương thì trả về False.

```python
def is_perfect(n):
    if n<1:
        return False
    sum=0
    for i in range(1,n):
        if n%i==0:
            sum+=i
    if sum==n:
        return True
n=int(input())
print(is_perfect(n))
```

Bài 31:Yêu cầu: Một chuỗi được gọi là perfect nếu nó chứa ít nhất một chữ hoa, ít nhất một chữ thường và ít nhất một chữ số.
Trả về True nếu chuỗi là perfect; ngược lại trả về False. Nếu đầu vào không phải chuỗi, hãy chuyển thành chuỗi

```python
def cpf(s):
    s=str(s)
    chuhoa=False
    chuthuong=False
    so=False
    for i in s:
        if i.isupper():
            chuhoa=True
        if i.islower():
            chuthuong=True
        if i.isdigit():
            so=True
    if chuhoa and chuthuong and so:
        return True
    else:
        return False
s=input()
print(cpf(s)) 
```

- Dùng các vòng lặp đơn giản và phép kiểm tra số nguyên.
- Không dùng tuple hoặc dictionary

```python
def primes_less_than(x):
    res=[]
    for i in range(2, x):
        is_prime=True
        for j in range(2,i):
            if i % j == 0:
                is_prime=False
                break
        if is_prime:
            res.append(i)
    return res
x=int(input())
print(primes_less_than(x))
```

Bài 38:Viết hàm inverted_triangle(n) trả về một tam giác ngược cân bằng gồm các dấu sao *, có độ rộng là n.
- n phải là một số nguyên dương lẻ, nếu không thì trả về chuỗi rỗng.
- Mỗi dòng có đúng n ký tự, số dấu sao giảm dần 2 đơn vị mỗi dòng cho đến khi còn 1 dấu sao

```python
def inverted_triangle(n):
    if n%2==0 and n<1:
        return ""
    rows=[]
    for i in range(n,0,-2):
        sokhoangtrong=(n-i)//2
        khoangtrong=" "*sokhoangtrong
        sao="*"*i
        row=khoangtrong+sao+khoangtrong
        rows.append(row)
    return "\n".join(rows)
n=int(input())
print(inverted_triangle(n))
```

- Không dùng tuple hoặc dictionary

```python
def sum_two_diagonals(mat):
    n=len(mat)
    tong=0
    for i in range(len(mat)):
        tong+=mat[i][i]
        if i!=n-1-i:
            tong+=mat[i][n-1-i]
    return tong
mat=[[1,2,3],[4,5,6],[7,8,9]]
print(sum_two_diagonals(mat))
```

Viết hàm column_with_max_sum(mat) trả về chỉ số bắt đầu từ 0 của cột có tổng lớn nhất.
Nếu có nhiều cột cùng tổng lớn nhất, trả về cột có chỉ số nhỏ nhất.
- Dùng vòng lặp để tính tổng từng cột.
- Không dùng tuple hoặc dictionary.

```python
def column_with_max_sum(ma_tran):
    if not ma_tran or not ma_tran[0]:
        return -1
    
    so_hang = len(ma_tran)
    so_cot = len(ma_tran[0])
    
    tong_lon_nhat = -float('inf')
    chi_so_cot_tot_nhat = 0
    
    for cot in range(so_cot):
        tong_cot_hien_tai = 0
        for hang in range(so_hang):
            tong_cot_hien_tai += ma_tran[hang][cot]
            
        if tong_cot_hien_tai > tong_lon_nhat:
            tong_lon_nhat = tong_cot_hien_tai
            chi_so_cot_tot_nhat = cot
            
    return chi_so_cot_tot_nhat
ma_tran=[[1,2,3],[4,5,6],[7,8,9]]
print(column_with_max_sum(ma_tran))
```

```python
Bài 43:Viết hàm zigzag_numbers(n) trả về một chuỗi biểu diễn ma trận n x n chứa các số từ 1 đến n^2. Các hàng cách nhau bởi ký tự xuống
dòng. Với chỉ số hàng bắt đầu từ 0: hàng chẵn đi từ trái sang phải, hàng lẻ đi từ phải sang trái. Các số trong cùng một hàng cách nhau bởi
một dấu cách
```

```python
def zigzag_numbers(n):
    mat=[]
    for i in range(n):
        row=list(range(i*n+1,(i+1)*n+1))
        if i%2!=0:
            row.reverse()
        mat.append(" ".join(map(str,row)))
    return "\n".join(mat)
print(zigzag_numbers(4))
```

Bài 44:Viết hàm count_solutions_3vars(a, b, c, t) trả về số nghiệm nguyên không âm (x, y, z) của phương trình a*x + b*y + c*z = t. Giả sử a, b, c,
t là các số nguyên. Nếu a <= 0 hoặc b <= 0 hoặc c <= 0 hoặc t < 0 thì trả về 0.


```python
def count_solutions_3vars(a, b, c, t):
    if a<=0 or b<=0 or c<=0 or t<0:
        return 0
    nghiem=0
    for x in range((t//a)+1):
        rutx=t-a*x
        for y in range((rutx//b)+1):
            ruty=rutx-b*y
            if ruty%c==0:
                nghiem+=1
    return nghiem
print(count_solutions_3vars(1, 2, 3, 5))
```

Bài 45:Số lần đổi chỗ kề nhau ít nhất để đưa số 1 duy nhất vào giữa ma trận có kích thước lẻ.
Bạn được cho ma trận n x n (với n là số lẻ). Ma trận chứa đúng một ô có giá trị 1,
còn tất cả các ô khác đều là 0. Trong một thao tác, bạn có thể đổi chỗ hai hàng kề nhau hoặc hai cột kề nhau.
Hãy tính số thao tác ít nhất để đưa số 1 vào đúng ô trung tâm

```python
def min_swaps_to_center(matrix, n):
    mid = n // 2
    r_idx, c_idx = -1, -1
    
    for i in range(n):
        for j in range(n):
            if matrix[i][j] == 1:
                r_idx, c_idx = i, j
                break
        if r_idx != -1: break
    
    return abs(r_idx - mid) + abs(c_idx - mid)
```

Yêu cầu thêm: Ưu tiên dùng map/dictionary.
Khai báo hàm: top3_frequent_names(names) -> list[str]

```python
def top3_frequent_names(names) -> list[str]:
    ten={}
    for i in names:
        if i in ten:
            ten[i]+=1
        else:
            ten[i]=1
    res=sorted(ten.keys(),key=ten.get,reverse=True)
    return res[:3]
names = ["An", "Bình", "An", "Chi", "Bình", "An", "Dũng"]
print(top3_frequent_names(names))
```

Bài 47:Yêu cầu: Cho danh sách các bộ ba (tuple gồm 3 số nguyên), hãy trả về bộ ba xuất hiện nhiều nhất.
Khai báo hàm: most_frequent_triple(triples) -> tuple[int,int,int]
- Phát sinh ValueError nếu danh sách đầu vào rỗng.


```python
def most_frequent_triple(triples) -> tuple[int,int,int]:
    if not triples:
        raise ValueError
    dem={}
    for i in triples:
        if i in dem:
            dem[i]+=1
        else:
            dem[i]=1
    res= max(dem,key=dem.get)
    return res
triples=[(1, 2, 3), (4, 5, 6), (1, 2, 3), (0, 0, 0)]
print(most_frequent_triple(triples))
```

Bài 48:Cho danh sách số nguyên nums và số nguyên k, hãy trả về True nếu tồn tại hai chỉ số khác nhau i và j sao cho:
- nums[i] == nums[j], và
- abs(i - j) <= k
Khai báo hàm: contains_nearby_duplicate(nums, k) -> bool


```python
def contains_nearby_duplicate(nums, k) -> bool:
   
    seen = set()
    for i, num in enumerate(nums):
        if num in seen:
            return True
        seen.add(num)
        if len(seen) > k:
            seen.remove(nums[i - k])
    return False
```

Bài 49:Bạn được cho hai mảng số nguyên 2 chiều items1 và items2.
Mỗi phần tử có dạng [value, weight], và value là duy nhất trong từng danh sách.
Yêu cầu: Trả về danh sách 2 chiều ret, trong đó mỗi phần tử là [value, total_weight],
với total_weight là tổng trọng lượng của cùng value trong cả hai danh sách.
Kết quả cần được sắp theo value tăng dần.

```python
def sum_items_by_value(items1, items2):
    counts = {}
    
    for value, weight in items1:
        counts[value] = counts.get(value, 0) + weight
        
    for value, weight in items2:
        counts[value] = counts.get(value, 0) + weight
        
    result = []
    for value in counts:
        result.append([value, counts[value]])
        
    result.sort()
    
    return result
```
