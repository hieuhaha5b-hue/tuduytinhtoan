### Bài 1: Tìm số lớn nhất của 1 list. KHONG DUNG HAM SAN CO CUA LIST

```python
lst = [10, 24, 5, 99, 50]
if not lst:
    max_val = None
else:
    max_val = lst[0]
    for num in lst:
        if num > max_val:
            max_val = num
print("Số lớn nhất:", max_val)
```

### Bài 2b: Tìm số nhỏ nhất của 1 list và vị trí của nó, nếu có nhiều số nhỏ nhất bằng nhau thì lấy chỉ số đầu tiên (nhỏ nhất)
10 0 0 12 34 5 6 7  8 89 0 1
0 1

```python

lst = [10, 0, 0, 12, 34, 5, 6, 7, 8, 89, 0, 1]
if not lst:
    result = None
else:
    min_val = lst[0]
    pos = 0
    for i in range(len(lst)):
        if lst[i] < min_val:
            min_val = lst[i]
            pos = i
    result = (min_val, pos)
print("Số nhỏ nhất và vị trí:", result)
```

### Bài 3: 
Tìm giá trị trung bình của các số nguyên dương trong một list.
Nếu không có số dương nào thì trả về "invalid"

```python
lst = [1, 2, 6, 1, -10, -10]
total = 0
count = 0
for num in lst:
    if num > 0:
        total += num
        count += 1

if count == 0:
    avg = "invalid"
else:
    avg = total / count
print("Trung bình số dương:", avg)
```

### Bài 4: Kiểm tra một list số nguyên có phải là đối xứng hay không
[1, 2, 3, 5, 3, 2, 1]

```python
lst = [1, 2, 3, 5, 3, 2, 1]
is_doi_xung = True
if not lst:
    is_doi_xung = None
else:
    n = len(lst)
    for i in range(n // 2):
        if lst[i] != lst[n - 1 - i]:
            is_doi_xung = False
            break
print("Đối xứng:", is_doi_xung)
```

### Bài 5: Đưa tất cả các số bằng 0 trong list về đầu của list. Các số khác không thay đổi thứ tự trước sau
3 4 6 7 1 9 0 0 23 0 0 34 67 89
0 0 0 0 3 4 6 7 1 9 23 34 67 89

### Bài 6: Xóa tất cả các số giống nhau trong list, giữ lại đúng 1 lần, Các số khác giữ nguyên vị trí trước 
[1, 2, 3, 4, 1, 2, 3, 4, 1, 1, 4, 3] 


```python
lst = [1, 2, 3, 4, 1, 2, 3, 4, 1, 1, 4, 3]
unique_lst = []
for num in lst:
    found = False
    for x in unique_lst:
        if x == num:
            found = True
            break
    if not found:
        unique_lst.append(num)
print("List duy nhất:", unique_lst)
```

### Bài 7: Kiểm tra một list số nguyên có phải là dãy tăng, dãy giảm hay vô danh
1 3 5 6 7 8 9: Tang

9 8 7 6 5 4 0: Giam

1 2 3 7 1 0 2: Vo danh

1 2 2 4 5 7: Vo danh

```python
lst = [1, 3, 5, 6, 7, 8, 9]
if len(lst) < 2:
    status = "Vo danh"
else:
    is_tang = True
    is_giam = True
    for i in range(len(lst) - 1):
        if lst[i] >= lst[i+1]: is_tang = False
        if lst[i] <= lst[i+1]: is_giam = False
    
    if is_tang: status = "Tang"
    elif is_giam: status = "Giam"
    else: status = "Vo danh"
print("Trạng thái dãy:", status)
```

### Bài 8: Kiểm tra một danh sách số nguyên có phải là ziczac khong
1 3 2 5 2 8 7 9 2 100: ziczac
34 2 56 7 90 20 30 7 120: ziczac
2 2 34 4 56 7 89 : không phải ziczac


```python
lst = [1, 3, 2, 5, 0, 7, 6, 8, 0]
is_zigzag = True
if len(lst) < 3:
    is_zigzag = False
else:
    for i in range(1, len(lst) - 1):
        check = (lst[i] > lst[i-1] and lst[i] > lst[i+1]) or \
                (lst[i] < lst[i-1] and lst[i] < lst[i+1])
        if not check:
            is_zigzag = False
            break
print("Dãy ziczac:", is_zigzag)
```

### Bài 9: Tìm và in ra tất cả các vị trí của số nguyên k trong một danh sách số nguyên cho trước. Nếu không có k trong ds thì in ra "-1".
VD: k = 3, [1,3,4,4,3,3,4,3,6,1] => k xuất hiện tại các vị trí (trả về 1 list) [1, 4, 5, 7]
k = 10, [1,3,4,4,3,3,4,3,6,1] => k không xuất hiện trong ds => trả về None

```python
lst = [1, 3, 4, 4, 3, 3, 4, 3, 6, 1]
k = 3
positions = []
for i in range(len(lst)):
    if lst[i] == k:
        positions.append(i)
if not positions: positions = None
print(f"Vị trí của {k}:", positions)
```

### Bài 11: 
Cho một danh sách các số nguyên có giá trị trong đoạn [0,9]. Liệt kê số lần xuất hiện của các số có mặt trong danh sách. Số nào xuất hiện trước thì liệt kê trước
VD 1 2 1 2 1 1 1 1

output: 
1: 6
2: 2


```python
lst = [2, 3, 8, 8, 8, 3, 1, 1, 2, 1, 4, 3, 4, 5]
counts = {}
for num in lst:
    if num in counts:
        counts[num] += 1
    else:
        counts[num] = 1
keys = list(counts.keys())
for i in range(len(keys)):
    for j in range(len(keys) - 1):
        if keys[j] > keys[j+1]:
            keys[j], keys[j+1] = keys[j+1], keys[j]

sorted_counts = {}
for k in keys:
    sorted_counts[k] = counts[k]
print("Số lần xuất hiện (đã sắp xếp):", sorted_counts)
```

### Bài 12: 
Cho một chuỗi kí tự chỉ gồm các kí tự chữ cái thường 'a'....'z'. Hãy liệt kê số lần xuất hiện của các kí tự có mặt trong chuỗi theo thứ tự tăng dần.
VD: "abcabcccc"
output:
a : 2
b : 2
c : 5

```python
s = "abcabcccc"
char_counts = {}
for char in s:
    if char in char_counts:
        char_counts[char] += 1
    else:
        char_counts[char] = 1

# Sắp xếp ký tự
char_keys = list(char_counts.keys())
for i in range(len(char_keys)):
    for j in range(len(char_keys) - 1):
        if char_keys[j] > char_keys[j+1]:
            char_keys[j], char_keys[j+1] = char_keys[j+1], char_keys[j]

for k in char_keys:
    print(f"{k} : {char_counts[k]}")
```
