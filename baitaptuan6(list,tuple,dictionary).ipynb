W6A12
[Tìm phần tử xuất hiện nhiều nhất] 
Nhập vào một list các số nguyên. Hãy tìm số xuất hiện nhiều lần nhất trong list. 
Nếu có nhiều số cùng có số lần xuất hiện nhiều nhất, hãy in ra số nhỏ nhất 
trong số đó

```python
s=list(map(int,input().split()))
so=[]
for x in s:
    so.append(int(x))
dem={}
nhieunhat=0
for i in so:
    if i in dem:
        dem[i]+=1
    else:
        dem[i]=1
for j in dem:
    if dem[j]>nhieunhat:
        nhieunhat=dem[j]
ss=[]
for k in dem:
    if dem[k]==nhieunhat:
        ss.append(k)
print(min(ss))



```

W6A21:Nhập một dòng gồm các chuỗi (tách bởi dấu cách). Tạo dictionary ánh xạ 
chuỗi -> tuple các vị trí (0-based) mà chuỗi đó xuất hiện

```python
data = input().split()
pos_dict = {}

for index, word in enumerate(data):
    if word in pos_dict:
        pos_dict[word].append(index)
    else:
        pos_dict[word] = [index]

for word in pos_dict:
    pos_dict[word] = tuple(pos_dict[word])

print(pos_dict)
```

W6A24:Viêết hàm đẩy tất cả các số 0 về đầu (về cuối), các số khác vẫn đúng vị trí trước sau: 0 2 3 2 4 0 0 0 9 => 0 0 0 2 3 2 4 9

```python
def l0(l):
    l1=[]
    l2=[]
    for i in l:
        if i==0:
            l1.append(i)
        else:
            l2.append(i)
    return l1+l2
l=list(map(int,input().split()))
print(l0(l))
```

W6A28:Cho một dãy các số nguyên, kiểm tra xem nó là dãy tăng thực sự hay giảm 
thực sự, hay vô danh (không tăng, không giảm)

```python
l=list(map(int,input().split()))
tangvohan=True
giamvohan=True
for i in range(len(l)-1):
    if l[i]>=l[i+1]:
        tangvohan=False
    if l[i]<=l[i+1]:
        giamvohan=False
if tangvohan:
    print("tangvohan")
elif giamvohan:
    print("giamvohan")
else:
    print("vodanh")
```
