chấm động, ngược lại trả về 'neither'. 
- Đầu vào: chuỗi s. 
- Đầu ra: một trong các giá trị {'int', 'float', 'neither'}. 
- Quy tắc: 
- Cho phép có dấu + hoặc - ở đầu. 
- Số nguyên phải chỉ gồm các chữ số và không có dấu chấm thập phân (ví dụ: '100'). 
- Số thực phải có dấu chấm và có chữ số ở cả hai phía của dấu chấm (ví dụ: '100.0', '-3.14'). 
- Các chuỗi như '.5', '5.', '1e3' được xem là 'neither' trong bài này.

```python
def detect_number_type(s):
    if s.isdigit():
        return "int"
    if "." in s:
        phan=s.split(".")
        if len(phan)==2 and phan[0].isdigit() and phan[1].isdigit():
            return "float"
    else:
        return "neither"
s=input()
print(detect_number_type(s))
    
```

B32:Yêu cầu: Viết hàm is_right_triangle(a, b, c) trả về True nếu a, b, c có thể là ba cạnh của một tam giác vuông, ngược lại trả

- Lưu ý: Kiểm tra bất đẳng thức tam giác và định lý Pythagore. Dùng sai số nhỏ khi so sánh số thực.

```python
def is_right_triangle(a,b,c):
    if a+b<c and a+c<b and b+c<a:
        return False
    else:
        if a*a+b*b==c*c or a*a+c*c==b*b or b*b+c*c==a*a:
            return True
is_right_triangle(3,4,5)
```

B33:Yêu cầu: Viết hàm compute(a, b, op) tính a (+|-||/) b tùy theo op. 
- Đầu vào: các số a, b và ký tự op thuộc {'+', '-', '', '/'}.

- Yêu cầu thêm: 
- Nếu op là '/' và b == 0, phát sinh ValueError. 
- Nếu op không hợp lệ, phát sinh ValueError

```python
def compute(a,b,op):
    if op=="+":
        res=a+b
    elif op=="-":
        res=a-b
    elif op=="/":
        if b==0:
            return "ValueError"
        else:
            res=a/b
    elif op=="":
        res=a*b
    else:
        return "ValueError"
    return res
a=float(input())
op=input()
b=int(input())

print(compute(a,b,op))
    
 
```

B35:Yêu cầu: Viết hàm divide_with_remainder(dividend, divisor) trả về bộ (q, r) trong đó q là thương nguyên và r là số dư sao 
cho dividend = divisorq + r với 0 <= |r| < |divisor| (dùng ngữ nghĩa của Python với // và %). 
- Đầu vào: số nguyên dividend, divisor (divisor khác 0). 
- Đầu ra: bộ hai số nguyên (q, r). 
- Ví dụ: divide_with_remainder(5, 3) -> (1, 2).

```python
def divide_with_remainder(sochia,sobichia):
    q=sochia//sobichia
    r=sochia%sobichia
    return (q,r)
divide_with_remainder(12,6)
```

B36:Yêu cầu: Viết hàm is_product_divisible_by_10(a, b) trả về True nếu a b chia hết cho 10, ngược lại trả về False. 
- Đầu vào: số nguyên a, b (có thể âm hoặc bằng 0). 
- Đầu ra: giá trị boolean. 
- Gợi ý: Một số chia hết cho 10 nếu chữ số tận cùng là 0 (tức là chia hết cho cả 2 và 5).

```python
def is_product_divisible_by_10(a,b):
    if (a*b)%10==0:
        return True
    else:
        return False
is_product_divisible_by_10(12,10)
```

B37:Viết hàm trả về tam giác vuông cân cao n bằng ký tự ''. 
Dòng thứ i (đánh số từ 1) có i dấu '', các dòng nối với nhau bằng ký tự xuống dòng. 
Ví dụ với n = 4

```python
def tamgiac(n):
    ket_qua = ""
    for i in range(1, n + 1):
       dong = '"' * i
       ket_qua += dong + "\n"
    return ket_qua
print(tamgiac(3))
```

B38:Viết hàm trả về chuỗi biểu diễn hình chữ nhật N x M bằng ký tự '', các hàng cách nhau bởi ký tự xuống dòng. 
Ví dụ, n=2, m=3

```python
def hcn(n,m):
    res=""
    for i in range(n):
        dong='"'*m
        res+=dong +"\n"
    return res
print(hcn(4,5))
```

B43:Viết hàm tính diện tích tam giác có ba cạnh a, b, c (dùng công thức Heron). Phát sinh ValueError nếu ba cạnh không thể 
tạo thành tam giác

```python
def stg(a,b,c):
    if a+b>c and a+c>b and b+c>a:
        p=(a+b+c)/2
        s=(p*(p-a)*(p-b)*(p-c))**0.5
        return s    
    else:
        return "ValueError"
stg(3,4,5)
stg(1,1,3)
```

B44:Viết hàm trả về tích các số nguyên từ 1 đến n (giai thừa). Với n=0, trả về 1.

```python
def gt(n):
    if n==0:
        giait=1
    else:    
        giait=1
        for i in range(1,n+1):
            giait*=i
    return giait
gt(3)
```

B45:Viết hàm trả về số lượng số nguyên dương k sao cho 1 <= k < n và k > (n - 1) / 2. (Ví dụ: n=9 -> 4).

```python
def hihi(n):
    dem=0
    for k in range(1,n):
        if k>(n-1)/2:
            dem+=1
    return dem
hihi(9)


```
