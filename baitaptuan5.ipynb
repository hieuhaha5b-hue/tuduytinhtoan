W5A4

```python
def sohoanhao(n):
    tong=0
    if n<1:
        return False
    for i in range(1,n):
        if n%i==0:
            tong+=i
    if tong==n:
        return True
    else:
        return False
n=int(input())
print(sohoanhao(n))
```

W5A7

```python
def maytinh(num1,num2,oprator):
    if oprator=='+':
        return num1+num2
    elif oprator=='-':
        return num1-num2
    elif oprator=='*':
        return num1*num2
    elif oprator=='/':
        if num2!=0:
            return num1/num2
        else:
            return "Lỗi: Không thể chia cho 0"
    else:
        return "Lỗi: Phép toán không hợp lệ"
num1=float(input())
oprator=input()
num2=float(input())
res=maytinh(num1,num2,oprator)
if type(res)==int or type(res)==float:
    print(round(res,2))
else:
    print(res)
    
```

W5A8

```python
def hamming(s1,s2):
    nhiphans1=bin(s1)[2:]
    nhiphans2=bin(s2)[2:]
    solonhon=max(len(nhiphans1),len(nhiphans2))
    nhiphans1=nhiphans1.zfill(solonhon)
    nhiphans2=nhiphans2.zfill(solonhon)
    khoangcach=0
    for i in range(solonhon):
        if nhiphans1[i]!=nhiphans2[i]:
            khoangcach+=1
    return khoangcach
s1=int(input())
s2=int(input())
print(hamming(s1,s2))
```

```python
def hamming(a,b):
    if len(a)!=len(b):
        return False
    adenb={}
    bdena={}
    for char_a,char_b in zip(a,b):
        if char_a in adenb:
            if adenb[char_a]!=char_b:
                return False
        else:
            adenb[char_a]=char_b
        if char_b in bdena:
            if bdena[char_b]!=char_a:
                return False
        else:
            bdena[char_b]=char_a
    return True
a=input()
b=input()
print(hamming(a,b))
```
