W7A3:Viết hàm thực hiện thuật 
toán bubble sort, Selection 
Sort để sắp xếp danh sách 
theo thứ tự tăng dần. Sắp 
xếp nổi bọt (Bubble Sort);

```python
def select_sort(arr):
    n = len(arr)
    for i in range(n):
        min_idx = i
        for j in range(i+1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr
arr=list(map(int, input().split()))
print(select_sort(arr))
print(bubble_sort(arr))
```

W7A7:Cho một danh sách các số 
nguyên nums, hãy tìm độ 
dài của dãy con tăng 
nghiêm ngặt dài nhất 
(không nhất thiết phải liền 
kề).

```python
def lengthOfLIS(nums):
    n = len(nums)
    dp = [1] * n
    for i in range(n):
        for j in range(i):
            if nums[i] > nums[j]:
                dp[i] = max(dp[i], dp[j] + 1)
    return max(dp)
nums = list(map(int, input().split()))
print(lengthOfLIS(nums))
```

W7A8:Bạn được cung cấp: một 
danh sách các khoảng thời 
gian interval = [[l1, r1], [l2, 
r2], ...] và một danh sách 
các truy vấn queries = [q1, 
q2, ...] 
Với mỗi truy vấn q, hãy tìm 
độ dài nhỏ nhất của 
khoảng thời gian [li, ri] sao 
cho li ≤ q ≤ ri. 
Nếu không tồn tại khoảng 
thời gian nào như vậy, hãy 
trả về -1 cho truy vấn đó. 
solve in < O(m.n), hint: 
binary search in the 
queries list

```python
import bisect

i = [[2, 8], [1, 5], [10, 15]]
q = [1, 4, 6, 11, 20]

n = len(q)
kq = [-1] * n

s_q = sorted((q[j], j) for j in range(n))
cs = [x[0] for x in s_q]

i.sort(key=lambda x: x[1] - x[0] + 1, reverse=True)

for l, r in i:
    v_l = bisect.bisect_left(cs, l)
    v_r = bisect.bisect_right(cs, r)
    
    d = r - l + 1
    for j in range(v_l, v_r):
        idx = s_q[j][1]
        kq[idx] = d

print(kq)
```

W7A10:Cho một danh sách các số 
nguyên nums và một số 
nguyên m, hãy tìm tổng 
mảng con lớn nhất theo 
modulo m. 
Tức là, với bất kỳ mảng 
con nums[i:j] nào, hãy tính 
sum(nums[i:j]) % m và trả 
về kết quả lớn nhất có thể.
Gợi ý: 
Duy trì tổng tiền tố (prefix 
sum) mod m và sử dụng 
tìm kiếm nhị phân để tìm 
tổng tiền tố nhỏ nhất chỉ 
lớn hơn tổng hiện tại một 
chút (để tối đa hóa hiệu số 
bao quanh).

```python
import bisect

nums = list(map(int, input().split()))
m = int(input())

prefix_sum = 0
max_mod_sum = 0
sorted_prefixes = [0]

for x in nums:
    prefix_sum = (prefix_sum + x) % m
    
    idx = bisect.bisect_right(sorted_prefixes, prefix_sum)
    
    if idx < len(sorted_prefixes):
        current_max = (prefix_sum - sorted_prefixes[idx] + m) % m
    else:
        current_max = prefix_sum
        
    if current_max > max_mod_sum:
        max_mod_sum = current_max
        
    bisect.insort(sorted_prefixes, prefix_sum)

print(max_mod_sum)
```

```python
n = int(input())

is_prime = [True] * (n + 1)
is_prime[0] = is_prime[1] = False
for p in range(2, int(n**0.5) + 1):
    if is_prime[p]:
        for i in range(p * p, n + 1, p):
            is_prime[i] = False

primes = []
for i in range(len(is_prime)):
    if is_prime[i]:
        primes.append(i)

max_len = 0
start_num = 0

for i in range(len(primes) - 1):
    p1 = primes[i]
    p2 = primes[i+1]
    current_len = p2 - p1 - 1
    
    if current_len > max_len:
        max_len = current_len
        start_num = p1 + 1

if max_len > 0:
    for i in range(start_num, start_num + max_len):
        print(i, end=" ")
    print("\n" + str(max_len))
else:
    print("Không có")
```

W7A29:Tìm Dãy con tăng dần liên tục dài nhất của một danh sách nhập vào từ 
bàn phím. Lấy dãy đầu tiên nếu có nhiều dãy có số phần tử bằng nhau

```python
a = list(map(int, input().split()))

if not a:
    print("Danh sách trống")
else:
    max_len = 1
    start_index = 0
    
    curr_len = 1
    curr_start = 0
    
    for i in range(1, len(a)):
        if a[i] > a[i-1]:
            curr_len += 1
        else:
            if curr_len > max_len:
                max_len = curr_len
                start_index = curr_start
            curr_len = 1
            curr_start = i
            
    if curr_len > max_len:
        max_len = curr_len
        start_index = curr_start
        
    result = a[start_index : start_index + max_len]
    print(*(result))
    print(max_len)
```

W7A30:Tìm Dãy con tăng dần KHÔNG liên tục dài nhất của một danh sách nhập 
vào từ bàn phím

```python
a = list(map(int, input().split()))

if not a:
    print("Danh sách trống")
else:
    n = len(a)
   
    L = [1] * n
  
    P = [-1] * n

    for i in range(1, n):
        for j in range(i):
            if a[i] > a[j] and L[i] < L[j] + 1:
                L[i] = L[j] + 1
                P[i] = j

    max_len = max(L)
    print(max_len)


    result = []
    curr_idx = L.index(max_len)
    while curr_idx != -1:
        result.append(a[curr_idx])
        curr_idx = P[curr_idx]
    
    result.reverse()
    print(*(result))
```
