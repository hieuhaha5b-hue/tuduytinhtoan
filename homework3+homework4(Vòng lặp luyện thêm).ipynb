```python
h = int(input())

for i in range(1, h + 1):
   
    if i == 1:
        print(" " * (h - i) + "*")
    
    elif i == h:
        print("*" * (2 * h - 1))
   
    else:
        khoang_trang_ngoai = " " * (h - i)
        khoang_trang_trong = " " * (2 * i - 3)
        print(f"{khoang_trang_ngoai}*{khoang_trang_trong}*")
```
